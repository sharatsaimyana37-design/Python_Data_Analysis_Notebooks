```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_us = df[df['job_country'] == 'United States'].copy()
df_us['job_posted_month'] = df['job_posted_date'].dt.strftime('%B')
df_us_pivot = df_us.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
df_us_pivot.reset_index(inplace=True)
df_us_pivot['month_no'] = pd.to_datetime(df_us_pivot['job_posted_month'], format='%B').dt.month
df_us_pivot.sort_values(by='month_no', inplace=True)
df_us_pivot.set_index('job_posted_month', inplace=True)
df_us_pivot.drop(columns='month_no', inplace=True)
```

```python
df_us_pivot
```

```python
df_us_software_pivot = pd.read_csv("https://lukeb.co/software_csv", index_col='job_posted_month')
df_us_software_pivot
```

```python
df_us_mergered = df_us_pivot.merge(df_us_software_pivot, on='job_posted_month')
df_us_mergered
```

```python
top_5 = df_us_mergered.sum().sort_values(ascending=False).head()
top_5 = top_5.index.tolist()
top_5
```

```python
df_us_mergered[top_5].plot(kind='line')
plt.title('Monthly Job Postings for the Tech Jobs in US')
plt.xlabel('2023')
plt.ylabel('Job_count')
plt.ylim(0, 20000)
plt.legend()
plt.show()
```
